**Netflix**! То, что началось в 1997 году как служба проката DVD, с тех пор превратилось в одну из крупнейших развлекательных и медиа-компаний.

Учитывая большое количество фильмов и сериалов, доступных на платформе, это прекрасная возможность проявить навыки работы с данными и погрузиться в индустрию развлечений. Мне был выдан CSV-файл, содержащий данные Netflix. Используя первоначальное данные, я изучите данные Netflix, чтобы выяснить, смогу ли я объяснить некоторые факторы, которые могут способствовать сокращению продолжительности фильмов.

Предоставленный набор данных `netflix_data.csv` имеет кололнки, подробно описанные ниже:

## Данные
### **netflix_data.csv**
| Колонка | Описание |
|--------|-------------|
| `show_id` | ID шоу |
| `type` | Тип шоу |
| `title` | Название шоу |
| `director` | Продюссер шоу |
| `cast` | Актерский состав |
| `country` | Страна происхождения |
| `date_added` | Дата добавления в Netflix |
| `release_year` | Год выхода шоу |
| `duration` | Продолжительность шоу |
| `description` | Описание шоу |
| `genre` | Жанр шоу |

```python
# Импорт библиотек
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Код!
net = pd.read_csv('netflix_data.csv')

netf = net[net['type']=='Movie']

cols = ["title", "country", "genre", "release_year", "duration"]
netflix_movies = netf[cols]
netflix_movies.head()

netflix_movies[netflix_movies['duration']<60]

colors=[]

for label, row in netflix_movies.iterrows():
    if row['genre'] == 'Children':
        colors.append("red")
    elif row['genre'] == 'Documentaries':
        colors.append("green")    
    elif row['genre'] == 'Stand-Up':
        colors.append("blue")
    else:
        colors.append("purple")
fig = plt.figure(figsize=(12,8))
plt.scatter(data=netflix_movies, x='release_year', y='duration', c=colors)
plt.title("Продолжительность фильма по году выхода")
plt.xlabel("Год выхода")
plt.ylabel("Продолжительность шоу в минутах")
plt.show()
```
